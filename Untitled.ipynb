```python
# %% 1. Импорты и настройки
import os
import re
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from PIL import Image
from transformers import BertTokenizer, BertModel
from sklearn.model_selection import GroupShuffleSplit
import matplotlib.pyplot as plt
import nltk

# Загружаем необходимые ресурсы NLTK
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

# Устройство для вычислений
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Используемое устройство:", DEVICE)

# Задайте корректные пути к данным
data_dir = "C:/Users/Deвайс/ML/to_upload"  # Корневая папка с CSV-файлами
train_images_dir = os.path.join(data_dir, "train_images")
test_images_dir = os.path.join(data_dir, "test_images")
```

```python
# %% 2. Загрузка данных
# Файл с описаниями для обучения
train_data = pd.read_csv(os.path.join(data_dir, "train_dataset.csv"))
# Файл с краудсорсинговыми оценками
crowd_annotations = pd.read_csv(
    os.path.join(data_dir, "CrowdAnnotations.tsv"),
    sep="\t", header=None, 
    names=["image_file", "query_id", "score", "yes", "no"]
)
# Файл с экспертными оценками
expert_annotations = pd.read_csv(
    os.path.join(data_dir, "ExpertAnnotations.tsv"),
    sep="\t", header=None,
    names=["image_file", "query_id", "ex1", "ex2", "ex3"]
)
# Файл с тестовыми запросами
test_queries = pd.read_csv(
    os.path.join(data_dir, "test_queries.csv"),
    sep="|", header=0,  # предполагается, что в файле есть заголовок
    names=["query_id", "query_text", "image"]
)
print("Train data shape:", train_data.shape)
print("Crowd annotations shape:", crowd_annotations.shape)
print("Expert annotations shape:", expert_annotations.shape)
print("Test queries shape:", test_queries.shape)
```

```python
# %% 3. Предобработка текстов и фильтрация изображений с юридически проблемным содержанием

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words]
    return " ".join(tokens)

# Применяем очистку текста к описаниям
train_data["clean_text"] = train_data["query_text"].apply(preprocess_text)

# Фильтруем записи, содержащие слова, связанные с несовершеннолетними
illegal_keywords = ["child", "baby", "kid", "toddler", "teenager"]

def contains_illegal(text):
    # Проверяем наличие ключевых слов (поиск по нижнему регистру)
    for word in illegal_keywords:
        if word in text.lower():
            return True
    return False

# Удаляем из обучающей выборки записи с упоминанием несовершеннолетних
train_data = train_data[~train_data["clean_text"].apply(contains_illegal)]
print("Train data после фильтрации по юридическим ограничениям:", train_data.shape)
```

```python
# %% 4. Агрегация оценок (экспертных и краудсорсинговых)

# Функция агрегации экспертных оценок: если все три оценки различны – исключаем запись (возвращаем None),
# иначе – выбираем оценку, получившую большинство голосов.
def aggregate_expert_scores(row):
    scores = [row["ex1"], row["ex2"], row["ex3"]]
    if len(set(scores)) == 3:
        return None
    else:
        return max(set(scores), key=scores.count)

expert_annotations["aggregated_score"] = expert_annotations.apply(aggregate_expert_scores, axis=1)
# Исключаем записи без консенсуса экспертов
expert_annotations = expert_annotations[expert_annotations["aggregated_score"].notnull()]

# Нормализуем экспертную оценку с 1-4 до 0-1 (формула: (score - 1) / 3)
expert_annotations["normalized_expert"] = (expert_annotations["aggregated_score"] - 1) / 3.0

# Для краудсорсинговых оценок – предположим, что столбец score уже представляет долю (если нет, можно нормировать)
if crowd_annotations["score"].max() > 1:
    crowd_annotations["normalized_crowd"] = crowd_annotations["score"] / crowd_annotations["score"].max()
else:
    crowd_annotations["normalized_crowd"] = crowd_annotations["score"]

# Объединяем оценки по идентификатору описания
merged_annotations = pd.merge(
    expert_annotations[["query_id", "normalized_expert"]],
    crowd_annotations[["query_id", "normalized_crowd"]],
    on="query_id", how="inner"
)

# Вычисляем итоговую оценку как взвешенное среднее (0.6 – эксперт, 0.4 – крауд)
merged_annotations["final_score"] = 0.6 * merged_annotations["normalized_expert"] + 0.4 * merged_annotations["normalized_crowd"]

# Объединяем итоговые оценки с данными описаний
train_df = pd.merge(train_data, merged_annotations[["query_id", "final_score"]], on="query_id", how="inner")
print("Итоговая обучающая выборка:", train_df.shape)

# Краткий анализ распределения целевой переменной
print(train_df["final_score"].describe())
train_df["final_score"].hist(bins=20)
plt.title("Распределение итоговых оценок")
plt.show()
```

```python
# %% 5. Векторизация изображений с помощью ResNet-18

# Выводим список колонок в train_data для проверки
print("Columns in train_data:", train_data.columns)

# Импортируем необходимые объекты для загрузки ResNet-18 с весами
from torchvision.models import resnet18, ResNet18_Weights

# Задаём веса для модели (это эквивалентно pretrained=True, но без предупреждений)
weights = ResNet18_Weights.IMAGENET1K_V1
resnet_model = resnet18(weights=weights)

# Убираем последний (полносвязный) слой
modules = list(resnet_model.children())[:-1]
resnet_model = nn.Sequential(*modules)
resnet_model.to(DEVICE)
resnet_model.eval()

# Определяем преобразования для изображений
image_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

def extract_image_features(image_path):
    try:
        image = Image.open(image_path).convert("RGB")
    except Exception as e:
        print(f"Ошибка загрузки изображения {image_path}: {e}")
        return None
    image_tensor = image_transform(image).unsqueeze(0).to(DEVICE)
    with torch.no_grad():
        features = resnet_model(image_tensor)
    features = features.view(features.size(0), -1)  # Приводим к виду (batch, features)
    return features.squeeze(0).cpu().numpy()

# Пример извлечения вектора изображения, используя столбец "image"
sample_image_path = os.path.join(train_images_dir, train_data.iloc[0]["image"])
feat = extract_image_features(sample_image_path)
if feat is not None:
    print("Размер вектора изображения:", feat.shape)
else:
    print("Не удалось извлечь вектор для изображения:", sample_image_path)
```

```python
# %% 6. Векторизация текстов с использованием модели BERT

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
bert_model = BertModel.from_pretrained("bert-base-uncased")
bert_model.to(DEVICE)
bert_model.eval()

def extract_text_features(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding="max_length", max_length=128)
    inputs = {k: v.to(DEVICE) for k, v in inputs.items()}
    with torch.no_grad():
        outputs = bert_model(**inputs)
    # Используем представление [CLS]-токена
    cls_embedding = outputs.last_hidden_state[:, 0, :]
    return cls_embedding.squeeze(0).cpu().numpy()

# Пример: извлечение вектора текста
txt_feat = extract_text_features(train_df.iloc[0]["clean_text"])
print("Размер текстового вектора:", txt_feat.shape)
```

```python
# %% 7. Подготовка обучающей выборки: вычисляем эмбеддинги и объединяем в один вектор
import os
import numpy as np
from tqdm import tqdm

# Пути к файлам для сохранения результатов (вы можете выбрать другие пути или названия)
features_file = os.path.join(data_dir, "train_features.npy")
labels_file = os.path.join(data_dir, "train_labels.npy")
groups_file = os.path.join(data_dir, "train_groups.npy")

if os.path.exists(features_file) and os.path.exists(labels_file) and os.path.exists(groups_file):
    print("Найдены сохранённые эмбеддинги. Загружаем их...")
    X = np.load(features_file)
    y = np.load(labels_file)
    groups = np.load(groups_file)
    print("Размер матрицы признаков:", X.shape)
else:
    print("Сохранённых эмбеддингов не найдено. Начинается вычисление...")
    features = []
    labels = []
    groups = []  # для последующего разбиения по группам (по image)
    
    for idx, row in tqdm(train_df.iterrows(), total=len(train_df), desc="Processing train examples"):
        # Используем столбец "image"
        img_path = os.path.join(train_images_dir, row["image"])
        img_feat = extract_image_features(img_path)
        if img_feat is None:
            continue  # пропускаем, если не удалось загрузить изображение
        txt_feat = extract_text_features(row["clean_text"])
        combined_feat = np.concatenate([img_feat, txt_feat])  # ожидаемый размер: 512 + 768 = 1280
        features.append(combined_feat)
        labels.append(row["final_score"])
        groups.append(row["image"])  # группируем по имени изображения
    
    X = np.array(features)
    y = np.array(labels).reshape(-1, 1)
    groups = np.array(groups)
    print("Размер матрицы признаков:", X.shape)
    
    # Сохраняем вычисленные данные на диск
    np.save(features_file, X)
    np.save(labels_file, y)
    np.save(groups_file, groups)
    print("Эмбеддинги сохранены.")
```

```python
# %% 8. Разбиение данных на тренировочную и тестовую выборки с использованием GroupShuffleSplit

gss = GroupShuffleSplit(n_splits=1, train_size=0.7, random_state=42)
train_idx, test_idx = next(gss.split(X, y, groups=groups))
X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]

print("Тренировочная выборка:", X_train.shape, y_train.shape)
print("Тестовая выборка:", X_test.shape, y_test.shape)

# Создаём PyTorch Dataset
class ImageTextDataset(Dataset):
    def __init__(self, features, targets):
        self.features = torch.tensor(features, dtype=torch.float32)
        self.targets = torch.tensor(targets, dtype=torch.float32)
    def __len__(self):
        return len(self.features)
    def __getitem__(self, idx):
        return self.features[idx], self.targets[idx]

train_dataset = ImageTextDataset(X_train, y_train)
test_dataset = ImageTextDataset(X_test, y_test)

batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
# %% 9. Определение и обучение модели предсказания соответствия
# Модель получает на вход объединённый вектор (1280 признаков) и выдаёт скалярное значение в диапазоне [0, 1]

class RegressionNet(nn.Module):
    def __init__(self, input_dim=1280):
        super(RegressionNet, self).__init__()
        self.fc1 = nn.Linear(input_dim, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, 128)
        self.fc4 = nn.Linear(128, 1)
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        # Применяем сигмоиду, чтобы выход был в диапазоне [0, 1]
        x = torch.sigmoid(self.fc4(x))
        return x

model = RegressionNet(input_dim=1280).to(DEVICE)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

num_epochs = 10
print("Начало обучения модели...")
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, targets in train_loader:
        inputs = inputs.to(DEVICE)
        targets = targets.to(DEVICE)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * inputs.size(0)
    epoch_loss = running_loss / len(train_dataset)
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}")

# Оценка модели на тестовой выборке
model.eval()
test_loss = 0.0
with torch.no_grad():
    for inputs, targets in test_loader:
        inputs = inputs.to(DEVICE)
        targets = targets.to(DEVICE)
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        test_loss += loss.item() * inputs.size(0)
test_loss /= len(test_dataset)
print("Test Loss:", test_loss)
```

```python
# %% 10. Тестирование модели: поиск наиболее релевантного изображения по текстовому запросу

# Для 10 случайных запросов из test_queries.csv получаем текстовый эмбеддинг,
# затем для каждого изображения из test_images вычисляем предсказанный score и выбираем изображение с максимальным значением.
sample_queries = test_queries.sample(10, random_state=42)

print("Тестирование на 10 случайных запросах из test_queries.csv...")
for idx, row in sample_queries.iterrows():
    query_text = row["query_text"]
    if contains_illegal(query_text):
        print("Запрос:", query_text)
        print("This image is unavailable in your country in compliance with local laws")
        continue
    # Применяем ту же предобработку текста
    query_text_clean = preprocess_text(query_text)
    query_text_features = extract_text_features(query_text_clean)
    
    image_scores = {}
    # Перебираем все изображения из папки test_images
    for image_name in os.listdir(test_images_dir):
        image_path = os.path.join(test_images_dir, image_name)
        img_feat = extract_image_features(image_path)
        if img_feat is None:
            continue
        # Объединяем вектор изображения и текстовый вектор запроса
        combined_feat = np.concatenate([img_feat, query_text_features])
        combined_feat_tensor = torch.tensor(combined_feat, dtype=torch.float32).unsqueeze(0).to(DEVICE)
        with torch.no_grad():
            score = model(combined_feat_tensor).item()
        image_scores[image_name] = score
    
    if not image_scores:
        print("Для запроса не найдено изображений в папке test_images.")
        continue

    best_image = max(image_scores, key=image_scores.get)
    best_score = image_scores[best_image]
    
    print(f"\nЗапрос: {query_text}")
    print(f"Наиболее релевантное изображение: {best_image} (score = {best_score:.4f})")
    
    # Вывод изображения с помощью matplotlib
    best_image_path = os.path.join(test_images_dir, best_image)
    try:
        img = Image.open(best_image_path)
        plt.imshow(img)
        plt.axis("off")
        plt.title(f"Score: {best_score:.4f}")
        plt.show()
    except Exception as e:
        print(f"Ошибка открытия изображения {best_image_path}: {e}")
```
